```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
import missingno as ms
import plotly.express as px
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",hue="sex",style="time",size="size")
```

```python
gap=px.data.gapminder()
temp_df=gap[gap["country"]=="India"]
temp_df
```

```python
sns.relplot(data=temp_df,x="year",y="lifeExp",kind="line")
```

```python
temp_df2=gap[gap["country"].isin(["India","Pakistan","China"])]
temp_df2
```

```python
sns.relplot(data=temp_df2,x="year",y="lifeExp",kind="line",hue="country")
```
